```python


datafile = '2023-03-31.LBNL.Bisabolene.vs.Hydrolysate.GB.txt'
df = pd.read_csv(datafile, sep='\t')
df = df.set_index('Hydrolysate')
map_this_column = "Bisabolene"
 
# Prepare a vector of color mapped to the 'Sugars' column
#my_palette = dict(zip(df.Sugars.unique(), ["red","orange","yellow","lightgreen","green","darkgreen","blue"]))
my_palette = dict(zip(df[map_this_column].unique(), ["#FF0000","#FF2000","#FF4000","#FF8000","#FFB000","#FFD000","#FFF000"]))
row_colors = df[map_this_column].map(my_palette)
 
# plot

sns.clustermap(df, metric="correlation", method="single", cmap="Purples", standard_scale=1) # Blues, etc.
plt.show()
```
